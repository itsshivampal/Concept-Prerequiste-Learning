```python
import pandas as pd
```

```python
df = pd.read_csv("data/predicted_prereq.csv")
df_prereq = pd.read_csv("data/first_prereq_pairs.csv")
```

```python
df_prereq.set_index("concept", inplace = True)
```

```python
upper_limit = 0.9
lower_limit = 0.12
```

```python
required_df = pd.DataFrame(columns = ["topic_a", "topic_b", "relation", "tfidf_score", "wiki_tfidf_score", "score"])

for i in range(df.shape[0]):
    topic_a = df["topic_a"].iloc[i]
    topic_b = df["topic_b"].iloc[i]
#     print(topic_a, topic_b)
    score = df_prereq.at[topic_a, topic_b]
    df.at[i, "score"] = score
    ground_truth = df["relation"].iloc[i]
    tfidf_score = df["tfidf_score"].iloc[i]
    wiki_tfidf_score = df["wiki_tfidf_score"].iloc[i]
    if ground_truth == 1 and tfidf_score == 0:
        required_df = required_df.append(df.iloc[i])
```

```python
required_df.reset_index(inplace = True)
required_df = required_df[["topic_a", "topic_b", "relation", "tfidf_score", "wiki_tfidf_score", "score"]]
```
